```python
import pandas as pd

import requests
from bs4 import BeautifulSoup as bs
```

```python
url="http://books.toscrape.com/"
response= requests.get(url)
response.status_code
```

```python
type(response.text)
```

```python
soup = bs(response.text)
type(soup)
```

```python
soup.find('title').text
```

```python
soup.find('title').text.strip()
```

```python
title= soup.find_all('h3')
title
```

```python
rating=soup.find_all('p')
rating
```

```python
price=soup.find_all('p', class_="price_color")
price
```

```python
stock= soup.find_all('p',class_='instock availability')
stock
```

```python
product_name=[]
for i in title:
  name=i.text
  product_name.append(name)

  product_price=[]
for i in price:
  price1=i.text
  product_price.append(price1)

  product_stock=[]
for i in stock:
  stock1=i.text.strip()
  product_stock.append(stock1)

table= pd.DataFrame  ({'Product_Name':product_name,'Price':product_price,'Stock':product_stock})
```
